```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
import pickle

# Load the CSV
csv = pd.read_csv('COUN_sep-feb 2023.csv', sep=',', header=0, encoding='utf-8')

# Split dataset into two parts based on the format
dfp1 = csv.iloc[:260497].copy()  # First part with m-d-Y format
dfp2 = csv.iloc[260498:].copy()  # Second part with d-m-Y format

# Clean and combine the data
def clean_and_combine_data(dfp1, dfp2):
    # Join date and time into a single column
    dfp1['Date Time'] = dfp1['Date'] + ' ' + dfp1['Time']
    dfp2['Date Time'] = dfp2['Date'] + ' ' + dfp2['Time']
    
    # Convert string to datetime format
    dfp1['Datetime'] = pd.to_datetime(dfp1['Date Time'], format='%m/%d/%Y %H:%M:%S')
    dfp2['Datetime'] = pd.to_datetime(dfp2['Date Time'], format='%d.%m.%Y %H:%M:%S')

    # Drop the temporary 'Date Time' column
    dfp1.drop('Date Time', axis=1, inplace=True)
    dfp2.drop('Date Time', axis=1, inplace=True)
    
    # Concatenate both parts and sort by datetime
    df = pd.concat([dfp1, dfp2])
    df = df.sort_values(by='Datetime')
    
    return df

df = clean_and_combine_data(dfp1, dfp2)

# Data Cleaning Function
def clean_data(df):
    print('\nCleaning data:')
    dfnumeric = df.copy()  # Copy for numeric data
    dfcleaned = df.copy()  # Copy for cleaned data
    
    for c in df.columns:
        if c in ['Date', 'Time', 'Datetime']:
            continue
        print(f'Cleaning {c}')
        
        mask = pd.to_numeric(df[c], errors='coerce').isna()
        dfcleaned[c].iloc[mask] = np.nan
        dfnumeric[c] = dfnumeric[c].astype(float)
        dfcleaned[c] = dfcleaned[c].astype(float)
        
        # Remove outliers based on Z-score
        z = stats.zscore(dfcleaned[c], nan_policy='omit')
        dfcleaned[c][np.abs(z) > 1.5] = np.nan
        dfcleaned[c] = dfcleaned[c].interpolate(method='linear')
    
    return dfnumeric, dfcleaned

dfnumeric, dfcleaned = clean_data(df)

# Plot original vs cleaned data
def original_clean_plot(df, dfc, var):
    plt.plot(df['Datetime'], df[var], label='Original')
    plt.plot(dfc['Datetime'], dfc[var], label='Cleaned')
    plt.legend()
    plt.title(var)
    plt.show()

original_clean_plot(dfnumeric, dfcleaned, 'Current')
original_clean_plot(dfnumeric, dfcleaned, 'Humidity outside')
original_clean_plot(dfnumeric, dfcleaned, 'Dew-point outside')

# Show statistical summary
def show_stats(df, var):
    print(f'\nStats for {var}:')
    print(f'Minimum: {np.min(df[var].astype(float))}, Maximum: {np.max(df[var].astype(float))}')
    
show_stats(dfnumeric, 'Temperature')
show_stats(dfcleaned, 'Temperature')
show_stats(dfcleaned, 'Dew-point outside')

# Perform PCA
dfpca = dfcleaned.drop(['Date', 'Time', 'Datetime', 'Temperature', 'Z'], axis=1)
dfpca = dfpca.dropna()

scaler = StandardScaler()
dfpcasc = scaler.fit_transform(dfpca)

pca = PCA(n_components=4)
comp = pca.fit_transform(dfpcasc)

# Plot PCA results
plt.scatter(comp[:, 0], comp[:, 1])
plt.title('Data distribution')
plt.xlabel('First Component')
plt.ylabel('Second Component')
plt.show()

# K-Means Clustering
kmeans = KMeans(n_clusters=2)
kmeans.fit(comp)

print('Centroids:\n', kmeans.cluster_centers_)

plt.scatter(comp[:, 0], comp[:, 1], c=kmeans.labels_)
plt.title('K-Means Clustering')
plt.show()

# DBSCAN Clustering
dbscan = DBSCAN(eps=1.1, min_samples=5)
chunk = comp[int(len(comp) / 3 * 0.9):]

dbscan.fit(chunk)

# Save DBSCAN model
with open('country-house-dbscan-2.bin', 'wb') as file:
    pickle.dump(dbscan, file)

# Load and predict with DBSCAN model
with open('country-house-dbscan-2.bin', 'rb') as file:
    dbscan_loaded = pickle.load(file)

print('Predictions:\n', dbscan_loaded.labels_)

plt.scatter(chunk[:, 0], chunk[:, 1], c=dbscan_loaded.labels_)
plt.title('DBSCAN Clustering')
plt.show()

# Cluster analysis
nclusters = set(dbscan_loaded.labels_)

print(f'Number of clusters formed: {len(nclusters)-1}')
for j in nclusters:
    print(f'Points in cluster {j}: {len(dbscan_loaded.labels_[dbscan_loaded.labels_ == j])}')
    
print(f'Out of clusters: {len(dbscan_loaded.labels_[dbscan_loaded.labels_ == -1])}')

# Statistical parameters per cluster
chunk_data = dfpca[int(len(dfpca) / 3 * 0.9):]

for j in nclusters:
    print(f'Cluster {j}:')
    cluster = chunk_data[dbscan_loaded.labels_ == j]
    print(cluster.describe())

# Optional: save model and predictions for future use
# with open('dbscan_model.pkl', 'wb') as f:
#     pickle.dump(dbscan, f)
```
